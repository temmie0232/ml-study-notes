# データの準備

```python
# データの準備
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(1000)  # 正規分布のランダムデータ
```

---
# 基本的なプロット関数

## 1. 線グラフ

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('線グラフの例')
plt.show()
```

## 2. 散布図

```python
plt.figure(figsize=(8, 4))
plt.scatter(x, y)
plt.title('散布図の例')
plt.show()
```


## 3. 棒グラフ

```python
plt.figure(figsize=(8, 4))
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
plt.bar(categories, values)
plt.title('棒グラフの例')
plt.show()
```


## 4. ヒストグラム

```python
plt.figure(figsize=(8, 4))
plt.hist(data, bins=30)
plt.title('ヒストグラムの例')
plt.show()
```

## 5. 箱ひげ図

```python
plt.figure(figsize=(8, 4))
plt.boxplot([np.random.normal(0, std, 100) for std in range(1, 4)])
plt.title('箱ひげ図の例')
plt.show()
```

## 6. 円グラフ

```python
plt.figure(figsize=(8, 8))
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('円グラフの例')
plt.show()
```

# プロットのスタイルオプション

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(12, 8))

# 異なる線のスタイル
plt.plot(x, y, 'r-', label='実線')
plt.plot(x, y+0.5, 'b--', label='破線')
plt.plot(x, y-0.5, 'g-.', label='一点鎖線')
plt.plot(x, y-1, 'k:', label='点線')

# 異なるマーカーのスタイル
plt.plot(x[::10], y[::10], 'ro', label='丸マーカー')
plt.plot(x[::10], y[::10]+0.5, 'bs', label='四角マーカー')
plt.plot(x[::10], y[::10]-0.5, 'g^', label='三角マーカー')

# 透明度と線の太さの設定
plt.plot(x, y-1.5, 'b-', linewidth=3, alpha=0.5, label='太い線+透明度')

plt.title('様々なプロットスタイル')
plt.legend()
plt.grid(True)
plt.show()
```

# サブプロット

```python
import matplotlib.pyplot as plt
import numpy as np

# 2x2のサブプロット
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# データの準備
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 1行1列目: 線グラフ
axes[0, 0].plot(x, y, 'b-')
axes[0, 0].set_title('線グラフ')
axes[0, 0].grid(True)

# 1行2列目: 散布図
axes[0, 1].scatter(x[::5], y[::5], c='r')
axes[0, 1].set_title('散布図')

# 2行1列目: ヒストグラム
axes[1, 0].hist(np.random.randn(1000), bins=30)
axes[1, 0].set_title('ヒストグラム')

# 2行2列目: バーグラフ
axes[1, 1].bar(['A', 'B', 'C', 'D'], [3, 7, 2, 4])
axes[1, 1].set_title('バーグラフ')

plt.tight_layout()
plt.show()
```
