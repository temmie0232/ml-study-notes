```python
import matplotlib.pyplot as plt
import numpy as np

# データの準備
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.exp(-x/5)

```

```python
# 図全体のサイズを設定
plt.figure(figsize=(12, 8))  # 横12インチ、縦8インチ

# 1. 基本的な線のスタイルとマーカーの組み合わせ
plt.plot(x, y1, 
         color='blue',           # 青色
         linestyle='--',         # 破線
         linewidth=2,            # 線の太さ
         marker='o',            # 丸いマーカー
         markersize=8,          # マーカーのサイズ
         markerfacecolor='red',  # マーカーの塗りつぶし色
         markeredgecolor='black',# マーカーの縁の色
         alpha=0.5,             # 透明度50%
         label='sin(x)')

# 2. 別のスタイルの線
plt.plot(x, y2,
         color='green',         # 緑色
         linestyle='-.',        # 一点鎖線
         linewidth=1.5,         # 線の太さ
         marker='^',           # 三角形のマーカー
         markersize=6,         # マーカーのサイズ
         alpha=0.7,            # 透明度30%
         label='cos(x)')

# 3. さらに別のスタイル
plt.plot(x, y3,
         color='red',          # 赤色
         linestyle=':',        # 点線
         linewidth=3,          # 線の太さ
         marker='s',          # 四角形のマーカー
         markersize=4,        # マーカーのサイズ
         alpha=0.6,           # 透明度40%
         label='sin(x)exp(-x/5)')

# グラフの設定
plt.title('さまざまなスタイルの線とマーカー', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)  # グリッド線の追加
plt.legend(fontsize=10)

# 軸の範囲設定
plt.xlim([0, 10])
plt.ylim([-1.5, 1.5])

# グラフの表示
plt.tight_layout()  # レイアウトの自動調整
plt.show()
```

```python
# 散布図の例
plt.figure(figsize=(10, 6))

# ランダムデータの生成
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 500  # マーカーサイズのバリエーション
colors = np.random.rand(50)       # カラーのバリエーション

# 散布図のプロット
plt.scatter(x, y,
           s=sizes,              # マーカーサイズ
           c=colors,             # カラー
           alpha=0.6,            # 透明度
           cmap='viridis',       # カラーマップ
           marker='o',           # マーカーの形
           edgecolors='black',   # マーカーの縁の色
           linewidth=1)          # マーカーの縁の太さ

# カラーバーの追加
plt.colorbar()

# グラフの設定
plt.title('散布図の例', fontsize=14)
plt.xlabel('x軸', fontsize=12)
plt.ylabel('y軸', fontsize=12)
plt.grid(True, alpha=0.3)

# グラフの表示
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# データの準備
x = np.linspace(0, 10, 100)  # 0から10までの100点
y1 = np.sin(x)               # sin波
y = np.random.rand(100)      # ランダムデータ
colors = np.random.rand(100) # カラー用のランダムデータ
sizes = 100 * np.random.rand(100)  # サイズ用のランダムデータ

# サブプロットの作成
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. 線グラフ
axes[0, 0].plot(x, y1, 'r--', linewidth=2, label='sin(x)')
axes[0, 0].set_title('線グラフ')
axes[0, 0].grid(True)
axes[0, 0].legend()

# 2. 散布図
scatter = axes[0, 1].scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
axes[0, 1].set_title('散布図')
fig.colorbar(scatter, ax=axes[0, 1])  # plt.colorbar() ではなく fig.colorbar() を使用

# 3. バーグラフ
x_bar = np.arange(5)
y_bar = np.random.rand(5)
axes[1, 0].bar(x_bar, y_bar, 
               alpha=0.7,
               color='skyblue',
               edgecolor='black',
               linewidth=1.5)
axes[1, 0].set_title('バーグラフ')

# 4. ステップグラフ
axes[1, 1].step(x[:20], y1[:20], 
                where='mid',
                color='green',
                linewidth=2,
                alpha=0.7)
axes[1, 1].set_title('ステップグラフ')
axes[1, 1].grid(True)

# レイアウトの調整
plt.tight_layout()
plt.show()
```
