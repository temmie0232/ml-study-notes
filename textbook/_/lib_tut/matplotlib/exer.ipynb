```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np
```

```python
# 0から10までの範囲で線グラフを作成してください。
# 要件:
# - X軸: 0から10までの100点
# - Y軸: sin(x)
# - タイトル、X軸ラベル、Y軸ラベルを設定
# - グリッド線を表示
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.figure(figsize=(6,3))

plt.plot(x,y)

plt.title("タイトル")
plt.xlabel("x軸")
plt.ylabel("y軸")

plt.grid(True)

plt.show()
```
